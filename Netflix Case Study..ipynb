```python
#In the director, replace the NaN values with unknown
df["director"].fillna("unknown",inplace=True)
```

```python
#In the country, replace the NaN values with unknown
df["country"].fillna("unknown",inplace=True)
```

```python
#drop the rest of null data

df.dropna(inplace=True)
```

```python
#count the values of movie and TV Show in the data
variety=df["type"].value_counts().reset_index()
variety
```

```python
#Show a graphical visual of the type

plt.figure(figsize=(10,5))
ax=sns.countplot(x="type",data=df)
plt.title("Count of Movies and TV Shows")
for p in ax.patches:
    height=p.get_height()
    ax.annotate(f"{height}",(p.get_x()+p.get_width()/2,height),ha="center")
```

```python
#Show a pie chart of the type

variety.plot.pie(y="count",autopct="%0.2f%%",labels=variety["type"])
```

```python
#count the values of director in the data and and create a dataframe as show_maker.

show_maker=df["director"].value_counts().reset_index()
show_maker
```

```python
#Show the values of the director data based on the type 

df.groupby(["director","type"]).size().reset_index(name="count")
```

```python
#Top 10 Show maker

df.groupby(["director","type"]).size().reset_index(name="count").nlargest(10,"count")
```

```python
#Top 10 Show maker ignoring the unkown column and create a dataframe as Top_10_show_maker.

Top_10_show_maker=df.groupby(["director","type"]).size().reset_index(name="count").nlargest(12,"count")[2:]
Top_10_show_maker
```

```python
#Plot a bar graph of the Top_10_Show_maker

plt.figure(figsize=(14,10))
ax=Top_10_show_maker.plot(x="director",y="count",kind="bar")
for p in ax.patches:
    height=p.get_height()
    ax.annotate(f"{height}",(p.get_x()+p.get_width()/2,height),ha="center")
```

```python
#get the top 10 movies with respect to country

top_10_movies_countries=df.groupby(["type","country"]).size().reset_index(name="count").nlargest(10,"count")
```

```python
top_10_movies_countries
```

```python
plt.figure(figsize=(10,8))
ax=top_10_movies_countries.plot(x="country",y="count",kind="bar")
for p in ax.patches:
    height=p.get_height()
    ax.annotate(f"{height}",(p.get_x()+p.get_width()/2,height),ha="center")
```
